```python
import yfinance as yf
stock_data = yf.download('AAPL', start='2010-01-01', end='2024-01-01')
stock_data.to_csv(r"C:\Users\Hp\Downloads\archive\infolimpioavanzadoTarget.csv")
print(stock_data.head())
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Users\Hp\Downloads\archive\infolimpioavanzadoTarget.csv", parse_dates=True)

# Remove missing values
df = df.dropna()

# Display first few rows
print(df.head())
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler

```

```python

# Load data
data = df[['Close']].values
```

```python
if data[0, 0] == 'AAPL':
  data = df[['Close']].values[1:]
```

```python
# Scale data between 0 and 1
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)
```

```python
# Prepare training data
X, y = [], []
for i in range(60, len(data_scaled)):
    X.append(data_scaled[i-60:i, 0])
    y.append(data_scaled[i, 0])

X, y = np.array(X), np.array(y)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))
```

```python
# Define LSTM model
model = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(X.shape[1], 1)),
    LSTM(units=50, return_sequences=False),
    Dense(units=25),
    Dense(units=1)
])
```

```python

# Compile model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python

# Train model
model.fit(X, y, batch_size=32, epochs=50)
```

```python
# Get last 60 days of stock data
last_60_days = data_scaled[-60:]
X_test = np.array([last_60_days])
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Predict next day's stock price
predicted_price = model.predict(X_test)
predicted_price = scaler.inverse_transform(predicted_price)
print(f"Predicted Stock Price: {predicted_price[0][0]}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df.index[-200:], df['Close'].values[-200:], label="Actual Prices")
plt.plot(df.index[-1], predicted_price[0][0], 'ro', label="Predicted Price")
plt.legend()
plt.title("Stock Price Prediction")
plt.show()
```
